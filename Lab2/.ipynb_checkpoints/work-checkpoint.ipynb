# Лабораторная 1. Построение графиков абсолютной погрешности различных методов.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib widget
```

```python
def function_1(x):
    return np.sin(np.power(x, 2))

def function_2(x):
    return np.cos(np.sin(x))

def function_3(x):
    return np.exp(np.sin(np.cos(x)))

def function_4(x):
    return np.log(x + 3)

def function_5(x):
    return np.power(x + 3, 0.5)
```

```python
def derivative_1(x):
    return 2 * x * np.cos(np.power(x, 2))

def derivative_2(x):
    return -np.cos(x) * np.sin(np.sin(x))

def derivative_3(x):
    return -np.cos(np.cos(x)) * np.sin(x) * np.exp(np.sin(np.cos(x)))

def derivative_4(x):
    return 1 / (x + 3)

def derivative_5(x):
    return 0.5 * np.power(x + 3, -0.5)
```

```python
def method_1(x, h, func):
    return (func(x + h) - func(x)) / h

def method_2(x, h, func):
    return (func(x) - func(x - h)) / h

def method_3(x, h, func):
    return (func(x + h) - func(x - h)) / (2*h)

def method_4(x, h, func):
    return 4 / 3 * (func(x +   h) - func(x -   h)) / (2*h) - \
           1 / 3 * (func(x + 2*h) - func(x - 2*h)) / (4*h)

def method_5(x, h, func):
    return 3 /  2 * (func(x +   h) - func(x -   h)) / (2*h) - \
           3 /  5 * (func(x + 2*h) - func(x - 2*h)) / (4*h) + \
           1 / 10 * (func(x + 3*h) - func(x - 3*h)) / (6*h)
```

```python
# Возвращает массив абсолютных ошибок соответсвующих h

def get_error(func, derivative_func, method, h, x_range):
    error = []
    for curr_h in h:
        error.append(np.sqrt(np.sum(np.power(derivative_func(x_range) - method(x_range, curr_h, func), 2)) / len(x_range)))
        
    return error
```

```python
def show_plot(h, func, derivative_func, func_name):
    x_0 = 0
    dy_0 = derivative_func(x_0)
        
    plt.figure(figsize=[8, 4])

    eps_1 = get_error(func, derivative_func, method_1, h, np.linspace(0, np.pi, 100))
    eps_2 = get_error(func, derivative_func, method_2, h, np.linspace(0, np.pi, 100))
    eps_3 = get_error(func, derivative_func, method_3, h, np.linspace(0, np.pi, 100))
    eps_4 = get_error(func, derivative_func, method_4, h, np.linspace(0, np.pi, 100))
    eps_5 = get_error(func, derivative_func, method_5, h, np.linspace(0, np.pi, 100))
    
    plt.plot(h, eps_1, label="Метод 1")
    plt.plot(h, eps_2, label="Метод 2")
    plt.plot(h, eps_3, label="Метод 3")
    plt.plot(h, eps_4, label="Метод 4")
    plt.plot(h, eps_5, label="Метод 5")
    
    plt.loglog()
    plt.xlabel("$h(n) = \\frac{2}{2^n}$")
    plt.ylabel("Погрешность")
    plt.title("Абсолютная погрешность дифференцирования функции: "+func_name)
    plt.grid()
    plt.legend()
    plt.show()
```

```python
n = np.arange(1, 22)
h_n = 2 / np.power(2, n)

print(n)
print(h_n)
```

```python
show_plot(h_n, function_1, derivative_1, "sin(x^2)")
```

```python
show_plot(h_n, function_2, derivative_2, "cos(sin(x))")
```

```python
show_plot(h_n, function_3, derivative_3, "exp(sin(cos(x)))")
```

```python
show_plot(h_n, function_4, derivative_4, "ln(x + 3)")
```

```python
show_plot(h_n, function_5, derivative_5, "(x + 3)^0.5")
```
