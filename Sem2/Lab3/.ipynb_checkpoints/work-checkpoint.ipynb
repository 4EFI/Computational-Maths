### Лабораторная работа 3. Численное решение периодической краевой задачи для линейного дифференциального уравнения

Требуется вычислить периодическое с периодом 1 решение уравнения:

$$y'' - yP^2(x) = f(x),$$

где $P^2(x) > 0$, $f(x)$ — заданные периодические функции, используя разностную схему второго порядка точности:

$$\begin{cases}
\frac{y_1 - 2y_0 + y_{N-1}}{h^2} - P^2(0)y_0 = f(0), \\
\frac{y_{n+1} - 2y_n + y_{n-1}}{h^2} - P^2(nh)y_n = f(nh), & n=1,2,...,N-2\\
\frac{y_0 - 2y_{N-1} + y_{N-2}}{h^2} - P^2(1-h)y_{N-1} = f(1-h),
\end{cases}$$

где $Nh = 1$. Для решения используется модификация метода прогонки (периодическая прогонка), так как стандартный метод не применим из-за периодических граничных условий.

**Параметры задачи:**
- Шаг сетки: $h = 0.005$
- Функции: $P^2(x) = 10 + \sin(2\pi x)$, $f(x) = \cos(2\pi x)$.

#### Описание метода

Для решения системы с периодическими граничными условиями применяется метод циклической прогонки, который состоит из следующих этапов:

1. **Прямой ход:** Вычисляются прогоночные коэффициенты $\alpha_n$, $\beta_n$, $\gamma_n$ для всех узлов сетки. Эти коэффициенты связывают значения решения в соседних узлах.

2. **Обратный ход:** Находятся коэффициенты $\mu_n$ и $\nu_n$, которые выражают решение в каждом узле через значение в начальном узле $y_0$.

3. **Определение $y_0$:** Используя периодическое условие $y_0 = y_N$, находится значение $y_0$.

4. **Вычисление решения:** По найденным коэффициентам и значению $y_0$ восстанавливается решение во всех узлах сетки.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Исходные функции
coeff_p = lambda x: 0
coeff_q = lambda x: -(10 + np.sin(2*np.pi*x))
rhs_f = lambda x: np.cos(2*np.pi*x)

def periodic_tridiagonal_solver(a_coeff, b_coeff, d_vect, step_size):
    num_points = len(d_vect)

    # Коэффициенты трехдиагональной матрицы
    lower_diag = 1 - 0.5 * a_coeff * step_size
    main_diag = 2 - b_coeff * step_size**2
    upper_diag = 1 + 0.5 * a_coeff * step_size
    rhs = d_vect * step_size**2

    # Прямой ход прогонки
    alpha = np.zeros(num_points)
    beta = np.zeros(num_points)
    gamma = np.zeros(num_points)

    alpha[1] = upper_diag[0]/main_diag[0]
    beta[1] = -rhs[0]/main_diag[0]
    gamma[1] = lower_diag[0]/main_diag[0]

    for i in range(1, num_points-1):
        denom = main_diag[i] - alpha[i] * lower_diag[i]
        alpha[i+1] = upper_diag[i] / denom
        beta[i+1] = (lower_diag[i] * beta[i] - rhs[i]) / denom
        gamma[i+1] = lower_diag[i] * gamma[i] / denom

    # Обратный ход для периодических условий
    mu = np.zeros(num_points)
    nu = np.zeros(num_points)

    last = num_points-1
    denom = lower_diag[last] * (alpha[last] + gamma[last]) - main_diag[last]
    mu[last] = -upper_diag[last] / denom
    nu[last] = (rhs[last] - lower_diag[last] * beta[last]) / denom

    for i in reversed(range(1, num_points)):
        mu[i-1] = alpha[i] * mu[i] + gamma[i] * mu[last]
        nu[i-1] = beta[i] + alpha[i] * nu[i] + gamma[i] * nu[last]

    # Вычисление решения
    solution = np.zeros(num_points)
    solution[0] = nu[0] / (1 - mu[0])
    solution[last] = mu[last] * solution[0] + nu[last]

    for i in reversed(range(1, num_points)):
        solution[i-1] = alpha[i] * solution[i] + beta[i] + gamma[i] * solution[last]

    return solution

def solve_periodic_bvp(p_func, q_func, f_func, step_size):
    domain_length = 1.0
    num_intervals = int(domain_length / step_size) + 1
    grid_points = np.linspace(0, 1, num_intervals)

    a_values = np.array([p_func(x) for x in grid_points])
    b_values = np.array([q_func(x) for x in grid_points])
    f_values = np.array([f_func(x) for x in grid_points])

    solution = periodic_tridiagonal_solver(a_values, b_values, f_values, step_size)

    return grid_points, solution

x_fine, y_fine = solve_periodic_bvp(coeff_p, coeff_q, rhs_f, 0.005)

# Построение графиков
plt.figure(figsize=(10, 6))
plt.plot(x_fine, y_fine, label=f'h = {0.005}', linewidth=1.5)
plt.xlabel('x')
plt.ylabel('Решение')
plt.title('Решение краевой задачи с периодическими условиями')
plt.grid(True)
plt.legend()
plt.show()
```
